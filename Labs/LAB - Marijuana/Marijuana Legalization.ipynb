# Marijuana legalization in the United States

The General Social Survey (GSS) is a biannual survey of American society. Every two years, researchers from the National Opinion Research Center (NORC) interview hundreds of American adults to investigate trends and changes in beliefs and attitudes.

The GSS added the question, "Do you think the use of marijuana should be made legal or not?" in 1973. When the question was first asked, 19% of Americans responded that marijuana should be legalized. But, a shift in attitudes has occurred over the last twenty years. In 1996, California was the first state to introduce legal marijuana use for treating serious medical conditions, such as cancer. In 2012, Colorado and Washington became the first US states to legalize recreational marijuana. Since then, 19 states have legalized recreational marijuana use, and 37 states have legalized medicinal use.

## Exploring support for marijuana legalization
Lawmakers and public policy researchers use data like the General Social Survey to explain relationships between personal beliefs and opinions. Ex: Are members of a certain political party more likely to support legalizing marijuana? Researchers analyzing data from the 2021 General Social Survey are interested in understanding whether a person supports legalizing marijuana based on three features: the respondent's age, education level, and political views.

## Explaining marijuana legalization with decision trees
Since researchers want a model that can be used to explain relationships, decision trees are a good choice. Decision tree models result in a visual representation of factors associated with a particular outcome. While marijuana legalization is complex, and a model based on three input features cannot explain a person's individual beliefs, the decision tree is useful for explaining population-level trends.

## Expanding the decision tree
Increasing the depth of a decision tree generally improves predictive performance. But, improved predictions come at a cost of decreased interpretability. If a model will be used to make predictions about new instances, such as whether a customer will make a purchase, a more complicated model is acceptable. If a model will be used to explain a phenomenon, such as factors affecting a voter's decisions, a simpler model may be preferred.

```python
# # Initialize a DecisionTreeClassifier with max_dpeth of 3, using seed as a random_state
classtreeModel = DecisionTreeClassifier(max_depth=3, random_state=seed)

# Fit the model with training data
classtreeModel.fit(X, y)
```

```python
# Resize the plotting window
plt.figure(figsize=[18, 8])

# Use tree.plot_tree to plot the decision tree.
p = tree.plot_tree(classtreeModel, filled=True, rounded=True, fontsize=14)
```

```python
# Use the tree model to predict y labesl
y_pred = classtreeModel.predict(X)
```

```python
# Display the Confusion matrix
print(metrics.confusion_matrix(y, y_pred))
```

## Feature importance
One disadvantage of decision trees is overfitting: The deeper the decision tree goes, the more likely that the tree is overfitted to the training data. Since the purpose of this decision tree is explanatory, and the tree is not intended to make new predictions, overfitting is less of a concern. But, a decision tree does not provide a measure of each input feature's relative importance.

Random forests are collections of many small decision trees. In a random forest, each decision tree is fit to a bootstrapped sample, resulting in a different model. Decision trees in a random forest randomly select a small subset of input features to test at each node vs. evaluate each feature at each node. Thus, the relative contribution or importance of each feature can be assessed based on how each feature contributes to the decision trees in a random forest.

<b>Feature importance</b> is a measure of the average impurity decrease within each decision tree of a random forest. A high feature importance means that a feature tended to perform better in the decision trees within the random forest.

```python
# Initialize the random forest model with max_depth=2, max_features=2
rfModel = RandomForestClassifier(max_depth=2, max_features=2, random_state=seed)

# Fit the random forest model with training data
rfModel.fit(X, y)
```

```python
importance = pd.DataFrame(
    data={
        'feature': rfModel.feature_names_in_,
        'importance': rfModel.feature_importances_,
    }
).sort_values('importance', ascending=False)

importance
```

```python
# Use sns.barplot to plot feature importance
sns.barplot(x='importance', y='feature', data=importance)
```
